# Univariate Analysis


## Price

```python
# check the distribution of the price

plt.figure(figsize=(12,10))
plt.hist(clean_data['price'], bins=50)
plt.title('Price distribution', fontdict = {'fontsize' : 22})
plt.show()
```

```python
# closer look at the distribution by plotting a simple histogram with 10 bins.

plt.figure(figsize=(20,10));  # makes the plot wider
plt.hist(clean_data['price'], color='r');  # plots a simple histogram
plt.axvline(clean_data['price'].mean(), color='m', linewidth=3,label='Mean')  # draw a line for mean
plt.axvline(clean_data['price'].median(), color='b', linestyle='dashed', linewidth=3,label='Medium') # draw a line for median
plt.axvline(clean_data['price'].mode()[0], color='g', linestyle='dashed', linewidth=3,label='Mode') # draw a line for mode

plt.title('Price for houses & apartment in Belgium', fontdict = {'fontsize' : 22})
plt.xlabel('Price')    # label the x-axis
plt.ylabel('Numbers of house or apartment')    # label the y-axis
plt.legend()      # Plot the legend

mean=round(clean_data['price'].mean())
median=round(clean_data['price'].median())
mode=round(clean_data['price'].mode())

print('\t\t\t\t\t\t Mean: ',mean,'  Median: ',median,'  Mode:',mode[0])
```

Note taken: 


*   In the above histogram we can see that the third bin with the price between 200,000 to 300,000 has the most numbers of observation.
*   Majority  of the observations lie within the second to fourth bin.  
*   Mean is 314,115, Median is 275,000 & mode is 249,000 euro.
